```python
import pandas as pd
```

```python
df = pd.read_csv('fraudTrain.csv')
df.head()
```

```python
df.info()
```

```python
# drop columns that won't be used in model
df.drop(columns=['Unnamed: 0', 'cc_num', 'merchant', 'first', 'last', 'street', 'city', 'state', 'zip', 'job','trans_num'], inplace=True)
```

```python
df.head()
```

```python
# check for missing values
df.isna().sum()
```

```python
df['category'].value_counts()
```

```python
# encode 'category' column and append, then drop 'category'
df = pd.concat([df, pd.get_dummies(df['category']).astype(int)], axis=1)
df.drop(columns=['category'], inplace=True)
df.head()
```

```python
# encode 'gender' column and append, then drop 'gender'
df = pd.concat([df, pd.get_dummies(df['gender']).astype(int)], axis=1)
df.drop(columns=['gender'], inplace=True)
```

```python
# convert dob and trans_date_trans_time to unixtime
df['dob'] = pd.to_datetime(df['dob']).apply(lambda x: x.timestamp())
df['trans_date_trans_time'] = pd.to_datetime(df['dob']).apply(lambda x: x.timestamp())
```
